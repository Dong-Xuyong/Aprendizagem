```python
name = "Leche Gansito Chocolate 4KG 200ml MLA 40680"
print(re.search("([A-Za-z\s]+)", name))
```

```python
print(re.search("\d+[gG]\s|\d+[mM][lL]\s", name))
```

```python
print(int(re.search("(\d+)[kK][gG]\s", name).group(1))*1000)
```

```python
print(re.search("([a-zA-Z]+)\s(\d+)$", name).group(1))
```

```python
for i in range(p_tab.shape[0]):
    p_lab = p_tab.iloc[i].NombreProducto
    desc, g, brand = [None] * 3
    
    descRE = re.search("([A-Za-z\s]+)", p_lab)
    if descRE:
        desc = descRE.group(1)
    gRE = re.search("(\d+)[gG]\s|(\d+)[mM][lL]\s", p_lab)
    if gRE:
        if gRE.group(1):
            g = gRE.group(1)
        else:
            g = gRE.group(2)
    else:
        kgRE = re.search("(\d+)[kK][gG]\s", p_lab)
        if kgRE:
            g = int(kgRE.group(1))*1000
    
    brandRE = re.search("([a-zA-Z]+)\s(\d+)$", p_lab).group(1)
    if brandRE:
        brand = brandRE
    for col, var in zip(newCols, [desc, g, brand]):
        
        p_tab.loc[i, col] = var
        
    if i%250==0:
        print(str(int(float(i)/p_tab.shape[0]*100)) + "% complete")
    
print('finished')

print(p_tab.shape)

print(p_tab.head(50))
```
